```python
#data preparation
df_copy = merge.copy()
```

```python
df_regg = df_copy[['Date', 'Qty']]
df_regg = df_regg.groupby('Date')[['Qty']].sum()
df_regg
```

```python
#melakukan split data 
train = df_regg.iloc[:-73]
test = df_regg.iloc[-73:]
```

```python
plt.figure(figsize=(16,8))
train['Qty'].plot(legend= True, label='Train')
test['Qty'].plot(legend = True, label='Test')
plt.show
```

```python
import pmdarima as pm

arima_model = pm.auto_arima(
    train, 
    seasonal=False, 
    stepwise=False, 
    suppress_warnings=True, 
    trace = True
)
arima_model.summary()
```

```python
#prediksi data test
prediction = arima_model.predict(len(test))
prediction
```

```python
train['Qty'].plot(legend = True, label = 'Training Data', figsize = (20, 10))
prediction.plot(legend=True, label = 'Predictions')
test['Qty'].plot(legend = True, label = 'Testing Data')
```

```python
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error, mean_squared_error
# Calculate MAPE (Mean Absolute Percentage Error)
def calculate_mape(actual, predicted) -> float:

    if not all([isinstance(actual, np.ndarray), isinstance(predicted, np.ndarray)]):
        actual, predicted = np.array(actual), np.array(predicted)

    # Calculate the MAPE value and return
    return round(np.mean(np.abs((actual - predicted) / actual)) * 100, 2)
def calculate_explained_variance(y_true, y_pred):
    mean_actual = np.mean(y_true)
    explained_variance = 1 - np.sum((y_true - y_pred) ** 2) / np.sum((y_true - mean_actual) ** 2)
    return explained_variance

mape = calculate_mape(test['Qty'], prediction)
mae = mean_absolute_error(test['Qty'], prediction)
mse = mean_squared_error(test['Qty'], prediction)
rmse = np.sqrt(mse)
explained_variance = calculate_explained_variance(test['Qty'], prediction)

print(f"MAPE: {mape:.2f}%")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"Explained Variance: {explained_variance:.2f}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
scaler.fit(df_cluster)
scaled_data = scaler.transform(df_cluster)
```

```python
inertia = []

for i in range(2,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
    
plt.figure(figsize=(10,6))
plt.xticks(list(range(2,11)))
plt.plot(range(2,11),inertia,marker='o')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans_3 = KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
kmeans_3.fit(scaled_data)
```

```python
df_cluster['cluster'] = kmeans_3.labels_
df_cluster
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(df_cluster['Qty'],
            df_cluster['TotalAmount'],
            c = df_cluster['cluster'])
plt.xlabel('Qty')
plt.ylabel('TotalAmount')
plt.title('Customer Segmentation using KMeans')
plt.show()
```
